# RSI(14) пересек 30 вниз - лонг лимитка на лой этой свечки. выход 4-16 свечек.

```python
from simple.binance import getHistMonth
from simple.plotly import chartFigure
from simple.pretty import pp, pmap, pxmap
import pandas as pd
import numpy as np
from talib import RSI
from functools import partial
from itertools import product
```

```python
# parameters
frame = '1h'
period = 14
threshold = 30
duration = 5
fee = 0.08
tickers = ['XLMUSDT', 'ETHUSDT', 'LTCUSDT', 'TRXUSDT', 'BNBUSDT', 'ADAUSDT',
           'BTCUSDT', 'ETCUSDT', 'XRPUSDT', 'XMRUSDT', 'ZECUSDT', 'EOSUSDT']
```

```python
def getOHLC(ticker: str) -> pd.Series:
    try:
        X = getHistMonth('2022-01', '2022-05', ticker, frame, spot=True)[['Open', 'High', 'Low', 'Close', 'Volume']]
        X.columns = pd.MultiIndex.from_product([[ticker], X.columns])
        return X
    except Exception as E:
        print(ticker, E)
```

```python
F = pd.concat(pmap(getOHLC, tickers), axis=1)
F
```

```python
def bt(ticker, period, threshold):
    C = F[ticker].copy()
    C['RSI'] = RSI(C.Close, period)

    def enterLong(row, threshold):
        return 1 if row[0] > threshold and row[1] < threshold else 0

    C['Target'] = C.RSI.rolling(window=2).apply(partial(enterLong, threshold=threshold)).shift()
    C['PrevLow'] = C.Low.shift()

    Pos = np.zeros(len(C))
    for index, row in enumerate(C.itertuples(index=False)):
        if row.Target > 0 and row.Low < row.PrevLow:
            Pos[index:index+duration] = 1

    C['Pos'] = Pos
    C['DPos'] = C.Pos.diff()
    C['Money'] = (-C.DPos * C.Low).cumsum()
    C['Fee'] = (C.DPos * C.Open * fee / 100).abs().cumsum()
    C['Balance'] = C.Money + C.Pos * C.Open
    return C


def getProfit(ticker, period, threshold):
    X = bt(ticker, period, threshold).Balance / F[ticker].Open[0] * 100
    X.name = ticker
    return X


def getProfitValue(ticker, period, threshold):
    X = bt(ticker, period, threshold).Balance / F[ticker].Open[0] * 100
    return X.iloc[-1]
```

```python
# grid search
Period = np.arange(4, 100, 2)
Threshold = np.arange(10, 90, 2)

P = pxmap(partial(getProfitValue, 'BTCUSDT'), product(Period, Threshold))
pp(P.pivot(index='period', columns='threshold', values='result').astype(int))
```

```python
# Mean equity chart
P = pd.concat(pmap(partial(getProfit, period=period, threshold=threshold), F.columns.get_level_values(0).unique()), axis=1)
P['Mean'] = P.mean(axis=1)
P
```

```python
D = {col: P[col] for col in P.columns}
D['Mean'] = dict(y=P.Mean, color='black', width=6)

chartFigure(**D)
```
