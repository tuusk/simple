```python
from simple.binance import getHistMonth
from simple.plotly import chartFigure
from simple.pretty import pp, pmap
import pandas as pd
import numpy as np
from talib import RSI
from functools import partial
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
frame = '1h'
tickers = ['XRPUSDT', 'BTCUSDT', 'LTCUSDT', 'BNBUSDT', 'ZECUSDT', 'ETHUSDT']
```

```python
def getOHLC(ticker: str) -> pd.Series:
    X = getHistMonth('2022-03', '2022-03', ticker, frame)[['Open', 'High', 'Low', 'Close', 'Volume']].iloc[:100]
    X.columns = pd.MultiIndex.from_product([[ticker], X.columns])
    return X
```

```python
F = pd.concat(pmap(getOHLC, tickers), axis=1)
F
```

```python
C = F.XRPUSDT.copy()
C['RSI'] = RSI(F.XRPUSDT.Close)
C
```

```python
fig = go.FigureWidget(make_subplots(rows=3, cols=1, shared_xaxes=True, vertical_spacing=0,
                                    row_heights=[0.7, 0.15, 0.15], specs=[[{"secondary_y": True}], [{}], [{}]]))

fig.update_layout(autosize=True, height=600, template='plotly_white',
                  legend=dict(x=0.4, y=1.01, orientation="h"),
                  margin=dict(l=45, r=5, b=10, t=10, pad=3),
                  xaxis_rangeslider_visible=False,
                  yaxis2_showgrid=False)

fig.add_candlestick(x=C.index,
    open=C.Open, high=C.High, low=C.Low, close=C.Close, name='OHLC', opacity=0.5)

fig.add_scattergl(name='RSI', x=C.index, y=C.RSI, row=2, col=1)

fig.update_xaxes(spikemode='across+marker', spikedash='dot', spikethickness=2, spikesnap='cursor')
fig.update_traces(xaxis='x1')

fig
```

```python
def threshold(C: pd.Series, th: float = 0.003) -> pd.Series:
    """Generate position based on delta threshold"""
    return np.where(C > th, -1, np.where(C < -th, 1, 0))
```

```python
Pos = Delta.apply(partial(threshold, th=0.002)).shift(1)
Pos
```

```python
ticker = 'BTCUSDT'
P = pd.DataFrame(Close[ticker])
P.columns = ['Price']
P['Delta'] = Delta[ticker]
P['Pos'] = Pos[ticker].fillna(0)
P['Amount'] = P.Pos.diff()
P['Cash'] = -(P.Amount * P.Price).cumsum()
P['Profit'] = P.Cash + P.Pos * P.Price
P
```
