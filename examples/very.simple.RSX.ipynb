# Chart with interactive oscillator on bottom panel

```python
import numpy as np
from simple.plotly import interactFigure
from simple.jurik import JRSX
```

```python
T = np.load('data/ETHUSDT.2022-05-08.tick.npz')['ETHUSDT']
len(T)
```

```python
def model(Period: int = (1000, 30000, 1000)):
    Tick = T['Price']
    RSX = JRSX(T['Price'], Period)
    return 0, locals()
```

```python
interactFigure(model, rows=2, height=750,
    Tick={'color': 'gray', 'opacity': 0.4},
    RSX={'color': 'green', 'opacity': 0.6, 'row': 2},
)
```
